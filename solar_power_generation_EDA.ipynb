PLOTTING

```python
# Plot histograms for each numerical column
for column in nc:
    plt.hist(df[column])
    plt.title(f"Histogram of {column}")
    plt.tight_layout()
    plt.show()
```

```python
# Create boxplots for numerical variables

for column in nc:
    plt.boxplot(df[column])
    plt.title(f"Boxplot of {column}")
    plt.tight_layout()
    plt.show()
```

```python
df['average-wind-speed-(period)']=df['average-wind-speed-(period)'].apply(lambda i:30 if i>30 else i)
```

```python
plt.boxplot(df['average-wind-speed-(period)'])
plt.show()
```

```python
df['average-pressure-(period)']=df['average-pressure-(period)'].apply(lambda i:28 if i>28  else i)
```

```python
df['power-generated']=df['power-generated'].apply(lambda i:31000 if i>31000 else i)
```

```python
plt.boxplot(df['power-generated'])
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle('Scatter Plots for Numerical_Columns', y=1.02)
plt.show()
```

```python
for nc in ['distance-to-solar-noon','temperature', 'wind-direction', 'wind-speed',
             'visibility', 'humidity', 'average-wind-speed-(period)',
            'average-pressure-(period)']:
    plt.figure()
    plt.scatter(df[nc], df['power-generated'])
    plt.xlabel(nc)
    plt.ylabel('power-generated')
    plt.title(f'Scatter plot of {nc} vs. power-generated')
    plt.show()

plt.figure()
plt.scatter(df.index, df['power-generated'])
plt.xlabel('Index')
plt.ylabel('power-generated')
plt.title('Scatter plot of Index vs. power-generated')
plt.tight_layout()
plt.show()
```

```python

# Create bar plots for each numerical column
for column in nc:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=30, kde=False)
    plt.title('Bar Graph: ' + column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# Create density plots for each numerical column
for column in nc:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df[column], fill=True)
    plt.title(f'Density Plot: {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.tight_layout()
    plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

Distance to Solar Noon: There is a strong negative correlation between the distance to solar noon and the power generated (-0.75). This suggests that power generation is highest around solar noon, when the sun is at its peak.

Humidity: Humidity also shows a moderate negative correlation with power generation (-0.52). Higher humidity levels seem to negatively affect power generation, possibly due to increased cloud cover or other atmospheric conditions that reduce sunlight.

Wind Speed: The average wind speed over the period has a moderate positive correlation with power generation (0.28). This indicates that wind speed may contribute positively to power generation, though its impact is less significant than the factors related to sunlight.

Temperature: Temperature has a slight positive correlation with power generation (0.13), indicating that higher temperatures might slightly enhance power generation, but the effect is relatively minor.

Other Variables: Variables like visibility, sky cover, and wind direction show weaker correlations with power generation, suggesting they have less impact on the amount of power generated.
